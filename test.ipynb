```python
import json

def parse_raw_cookie_string(raw_string, domain=".divyabhaskar.co.in"):
    """Converts a raw HTTP cookie string into Playwright's required JSON list."""
    playwright_cookies = []
    
    # Split the string by the semicolon delimiter
    cookie_pairs = raw_string.split(';')
    
    for pair in cookie_pairs:
        # Strip whitespace and split by the first equals sign
        pair = pair.strip()
        if not pair:
            continue
            
        parts = pair.split('=', 1)
        if len(parts) == 2:
            name, value = parts
            playwright_cookies.append({
                "name": name,
                "value": value,
                "domain": domain,
                "path": "/"
            })
            
    return playwright_cookies

# Example Usage:
raw_cookies = "_gid=GA1.3.1336827284.1776999660; dbskrrcr=1; dbskrat=Gu3m3gVhYiR8d_iSCfG8yj7TJeyZ0ZhbQvhpD_MCTbH2c6iZx5xtgTO5YLeck0MKgwY4T00ijnfHFI1TPaV73aeb5YHVItZBepN3L25CNoyDzI_ZbCat2YdvkG3JdWM_zg5mg1xnoG9bTyp-cHhAr5SY6xsRZRAbQbRXkIoHGa5NUxrODHv0Ld9STW--dzZG9WU_1wArH4VKmjZ5lxmlJXEgHJIXoRPCfc6C3IMlaxzGktRUmVsmQNCSLUyR9qcH9A5mRK4vYfEV6y_HPufc9mUZG_5rmKxMN09ujDoLTFU=; dbskrrt=01cd50af125647ce926b466ffe2fdf6a; dbskruid=2047511152554094592; pt=eyJhbGciOiJIUzI1NiJ9.eyJpYXQiOjE3NzcwMDU2NDd9.kNvAvKQm5kybxWHXeevSv0e9oM6Gfid1piF7uAYa4VQ; _ga_95X6FXVWS3=GS2.1.s1777005795$o2$g1$t1777005795$j60$l0$h0; AMP_TOKEN=%24NOT_FOUND; _ga=GA1.3.1290288046.1776999660" 
formatted_cookies = parse_raw_cookie_string(raw_cookies)

# You can then pass this list directly to context.add_cookies()
# or save it to your db_session.json file:
with open('db_session.json', 'w') as f:
    json.dump(formatted_cookies, f, indent=2)
```
